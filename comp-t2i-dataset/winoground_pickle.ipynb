```python
winogrand_caption_dir = '/home/jasonlin/repos/datasets/t2i_benchmark/winoground/examples.jsonl'

import jsonlines 

ex_data = {}
ex_id_pairs = []

with jsonlines.open(winogrand_caption_dir) as reader:
    for obj in reader:
        print(obj)
        if obj['id'] == 399:
            continue
        ex_data[obj['image_0']] = {0: {'text': obj['caption_0'], 'heldout_pairs': []}}
        ex_data[obj['image_1']] = {0: {'text': obj['caption_1'], 'heldout_pairs': []}}
        # construct pair list for train/test split
        ex_id_pairs.append((obj['image_0'], obj['image_1']))
```

```python
import numpy as np

np.random.shuffle(ex_id_pairs)
ex_id_pairs
```

```python
TRAIN_SPLIT = 0.75
num_sample_pairs = len(ex_id_pairs)
num_train = int(num_sample_pairs * TRAIN_SPLIT)

train_pairs, test_pairs = ex_id_pairs[:num_train], ex_id_pairs[num_train:]
```

```python
test_img_list = [img for p in test_pairs for img in p]
train_img_list = [img for p in train_pairs for img in p]
```

```python
ex_split = {'train': train_img_list, 'test_seen': test_img_list, 'all': [img for p in ex_id_pairs for img in p]}
```

```python
with open('/home/jasonlin/repos/datasets/t2i_benchmark/winoground/pickle/data.pkl', 'wb') as data_file:
    pickle.dump(ex_data, data_file)
    
with open('/home/jasonlin/repos/datasets/t2i_benchmark/winoground/pickle/split.pkl', 'wb') as split_file:
    pickle.dump(ex_split, split_file)
```
